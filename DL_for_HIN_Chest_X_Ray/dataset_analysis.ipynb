```python
import pandas as pd
from collections import defaultdict
```

```python
# Load your CSV file
csv_file_path = 'HIN_archive/Data_Entry_2017.csv'
df = pd.read_csv(csv_file_path)

unique_labels = df["Finding Labels"].str.split("|").explode().unique()
```

```python
# Initialize an empty dictionary to hold label combinations
label_combinations = defaultdict(lambda: defaultdict(int))

# Process each row in the dataset
for _, row in df.iterrows():
    # Get the labels for this entry (split by '|')
    labels = row["Finding Labels"].split('|')
    labels = sorted(labels)
    labels_string = '_'.join(labels)
    
    if len(labels) == 1:
        label_combinations[labels[0]]['alone'] += 1
        continue

    # For each label, count combinations with other labels
    for label in labels:
        for other_label in labels:
            if label != other_label:
                # Sort the labels to maintain consistent ordering
                label_combinations[label][labels_string] += 1

# Print the result
for label, combinations in label_combinations.items():
    print(f"Label: {label}")
    for combination, count in combinations.items():
        print(f"  {combination}: {count}")
```

```python
# Initialize an empty dictionary to hold label combinations
label_combinations = defaultdict(lambda: defaultdict(int))

# Process each row in the dataset
for _, row in df.iterrows():
    # Get the labels for this entry (split by '|')
    labels = row["Finding Labels"].split('|')
    
    if len(labels) == 1:
        label_combinations[labels[0]]['alone'] += 1
        continue

    # For each label, count combinations with other labels
    for label in labels:
        for other_label in labels:
            if label != other_label:
                # Sort the labels to maintain consistent ordering
                combination = tuple(sorted([label, other_label]))
                label_combinations[label][combination] += 1

# Print the result
for label, combinations in label_combinations.items():
    print(f"Label: {label}")
    for combination, count in combinations.items():
        print(f"  {combination}: {count}")
```

```python
label_combinations_items = sorted(label_combinations.items(), key=lambda item: item[1]['alone'])
total_alone = 0

# Calculate the total occurrences for each label
label_totals = {label: sum(count for combination, count in combinations.items()) for label, combinations in label_combinations.items()}

# Print the result in percentages
for label, combinations in label_combinations_items:
    total = label_totals[label]
    print(f"Label: {label} with total of {total}")
    for combination, count in combinations.items():
        # Calculate percentage for each combination
        if combination == 'alone':
            total_alone += count
            percentage = (count / total) * 100
            print(f"  Alone: {percentage:.2f}%")
        else:
            percentage = (count / total) * 100
            print(f"  {combination}: {percentage:.2f}%")
print(total_alone)
```

```python
import cv2
IMAGE_DIR = "HIN_archive/images"

image = cv2.imread(IMAGE_DIR + '/00014716_007.png', cv2.IMREAD_GRAYSCALE)
x, y, w, h = 686, 131, 185, 313

# image = cv2.imread(IMAGE_DIR + '/00019271_064.png', cv2.IMREAD_GRAYSCALE)
# x, y, w, h = 265, 509, 593, 268


image_bgr = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
overlay = image_bgr.copy()
color = (0, 0, 255)  # Green in BGR
alpha = 1
cv2.rectangle(overlay, (x, y), (x + w, y + h), color, 7)  # -1 fills the rectangle
cv2.addWeighted(overlay, alpha, image_bgr, 1 - alpha, 0, image_bgr)


cv2.imwrite("xray_50.png", cv2.resize(image_bgr, (50, 50)))
cv2.imwrite("xray_200.png", cv2.resize(image_bgr, (200, 200)))
cv2.imwrite("xray_400.png", cv2.resize(image_bgr, (400, 400)))
cv2.imwrite("xray_1024.png", image_bgr)
```
